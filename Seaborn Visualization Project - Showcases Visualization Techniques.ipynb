```python
### Importing the necessary libraries
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
### Setting up the df and looking over the data
df = pd.read_csv("WorldCupMatchesReal.csv")
print(df.head())
print(df.info())
```

```python
### Creating the Total Goals Column
total_goals = df["Home Team Goals"] + df["Away Team Goals"]
df["Total Goals"] = total_goals
```

```python
### Setting the Seaborn style
sns.set_style("whitegrid")
sns.set_context("poster")
```

```python
### Creating the figure
f, ax = plt.subplots(figsize=(20, 7))

ax = sns.barplot(data=df, x='Year', y='Total Goals')
plt.title("Goals Scored in the World Cup")
plt.show()
```
